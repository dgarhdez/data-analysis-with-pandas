### Operations with arrays: element-wise operations

```python
# creating vectors
array1 = np.array([3, 5, 7, 9])
array2 = np.array([1, 5, 9, 13])
```

```python
# addition
array1 + array2
```

```python
# substraction
array1 - array2
```

```python
# multiplication
array1 * array2
```

```python
# division
array1 / array2
```

```python
# multiplication by scalar
scalar1 = 3.755558

array1 * scalar1
```

```python
# power
array1 ** 3.6
```

```python
# more power
np.power(array1, 3.6)
```

```python
# logical operations
array1 = np.array(range(10))
array1[array1>5]
```

```python
# more logical operations

array1[array1%2==0]  # even elements
```

```python
array1%2==0
```

```python
# np.where
# np.where(logical_condition, value_if_true, value_if_false)

array1 = np.arange(1, 15, 3)
print(array1)

np.where(array1%2==0, "even", "odd")
```

### Operations with arrays: rearraging elements

```python
# create a 2D array (matrix)

mtx = np.array([
    [1, 2],
    [0, 0],
    [1, 0]
])
```

```python
# mtx reshape
# the key to reshape is to keep the product of (rows*columns) constant

np.reshape(mtx, (2, 3))
```

```python
# more mtx reshape

np.reshape(mtx, (1, 6))
```

```python
# mtx flatten
# similar to np.reshape(1, n) but this returns a copy - doesn't change the original value

mtx.flatten()
```

```python
# transposing a square matrix

# let's create a square matrix
sq_mtx = np.array([
    ["r1c1", "r1c2", "r1c3"],
    ["r2c1", "r2c2", "r2c3"],
    ["r3c1", "r3c2", "r3c3"]
])
```

```python
# using transpose()

sq_mtx.transpose()
```

### Basic algebra operations

```python
# dot product of two vectors

v1 = np.array([1, 2, 7])
v2 = np.array([8, 4, 2])

np.dot(v1, v2)  # 8*1 + 4*2 + 7*2
```

```python
# Multiplying matrix and vector

v1 = np.array([1, 0 ,1])
mtx1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

np.dot(v1, mtx1)  #
```

```python
# always check dimensions: (n, m) * (m, p) = (n, p)

np.dot(mtx1, v1)
```

```python
# also with @ operator
mtx1 @  v1
```

```python
# multiplication of two square matrices
msq1 = np.array([
    [1, 0, 1],
    [2, 5, 4],
    [3, 6, 1]
])

msq2 = np.array([
    [5, 0, 3],
    [5, 1, 4],
    [0, 6, 8]
])

msq1 @ msq2
```

```python
# inverse of a matrix
inv_msq1 = np.linalg.inv(msq1)
```

```python
print(msq1)
print(inv_msq1)
```

```python
# multiply matrix by its inverse equals an identity matrix!

print(msq1 @ inv_msq1)
```
